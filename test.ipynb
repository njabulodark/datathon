```python
def get_data(year, limit):
    #get the tweets from the year
    tweets_list2 = []
    for i,tweet in enumerate(sntwitter.TwitterHashtagScraper(f'corruption since:{year}-01-01 until:{year+1}-01-01').get_items()):
        if i>1200:
            break

        if "south africa" in tweet.user.location.lower():
            tw = tweet.content
            tweets_list2.append([tweet.hashtags])

    #make a dataframe
    hash_data = pd.DataFrame(tweets_list2, columns=['Hashtags'])

    #make a list of all the hashtags
    def make_list(stri):
        lis = []
        stri = stri.split(',')
        for i in range(len(stri)):
            lis.append(stri[i][2:-1])
        return lis

    hash_data["Hashtags"] = hash_data["Hashtags"].agg(make_list)

    #combine all the lists into one list
    lis = []
    for i in range(len(df)):
        for j in range(len(df["Hashtags"][i])):
            lis.append(df["Hashtags"][i][j])

    #count the number of times each hashtag appears
    from collections import Counter
    Counter(lis)

    #get a list top 3 hashtags
    Counter(lis).most_common(3) 

    #dic to list


    # Creating list to append tweet data to
    tweets_list1 = []
    tweets_list2 = []
    tweets_list3 = []
    
    for i,tweet in enumerate(sntwitter.TwitterHashtagScraper(f'#{Counter(lis).most_common(3)[0][0]} since:2017-01-01 until:2018-01-01').get_items()):
        if i>limit:
            break
        tweets_list1.append([tweet.date.strftime("%y:%m:%d %H:%M:%S"), tweet.content, tweet.user.username, tweet.retweetCount, tweet.user.location, tweet.hashtags])
    
    for i,tweet in enumerate(sntwitter.TwitterHashtagScraper(f'#{Counter(lis).most_common(3)[1][0]} since:2017-01-01 until:2018-01-01').get_items()):
        if i>limit:
            break
        tweets_list2.append([tweet.date.strftime("%y:%m:%d %H:%M:%S"), tweet.content, tweet.user.username, tweet.retweetCount, tweet.user.location, tweet.hashtags])
    
    for i,tweet in enumerate(sntwitter.TwitterHashtagScraper(f'{Counter(lis).most_common(3)[2][0]}since:2017-01-01 until:2018-01-01').get_items()):
        if i>limit:
            break
        tweets_list3.append([tweet.date.strftime("%y:%m:%d %H:%M:%S"), tweet.content, tweet.user.username, tweet.retweetCount, tweet.user.location, tweet.hashtags])

    # Creating a dataframe from the tweets list above
    tweets_df = pd.DataFrame(tweets_list1 + tweets_list2 + tweets_list3, columns=['Datetime', 'Text', 'Username', 'Retweets', 'Location', 'Hashtags'])
    tweets_df.to_csv(f'{year}data.csv', index=False)
    return tweets_df

get_data(2017, 100)

    
        
```
