```python
import pandas as pd
import numpy as np
import snscrape.modules.twitter as sntwitter
import re
```

```python
# Creating list to append tweet data to
tweets_list = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#RSA').get_items()):
    if i>10:
        break
    tweets_list.append([tweet.date.strftime("%y:%m:%d %H:%M:%S"), tweet.user.username, tweet.content, tweet.retweetCount, tweet.replyCount, tweet.likeCount, tweet.quoteCount, tweet.lang, tweet.user.location, tweet.hashtags])
```

```python
df = pd.DataFrame(tweets_list, columns=['Datetime', 'Username', 'Text', 'Retweets', 'Replies', 'Likes', 'Quotes', 'Language', 'Location', 'Hashtags'])
df
```

```python
def clean_tweet(tweet):
    tweet = re.sub(r'@[A-Za-z0–9]+', '', tweet) #Removing @mentions
    tweet = re.sub(r'#', '', tweet) # Removing '#' hash tag
    tweet = re.sub(r'RT[\s]+', '', tweet) # Removing RT
    tweet = re.sub(r'https?:\/\/\S+', '', tweet) # Removing hyperlink
    return tweet
```
