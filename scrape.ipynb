```python
import pandas as pd
import numpy as np
import snscrape.modules.twitter as sntwitter
```

```python
# Creating list to append tweet data to
tweets_list = []

for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#corruption').get_items()):
    if i>1000:
        break
    tweets_list.append([tweet.date, tweet.user.username, tweet.content, tweet.retweetCount, tweet.replyCount, tweet.likeCount, tweet.quoteCount, tweet.lang, ])
```

```python
df = pd.DataFrame(tweets_list, columns=['Datetime', 'Username', 'Text', 'Retweets', 'Replies', 'Likes', 'Quotes', 'Language'])
df
```
