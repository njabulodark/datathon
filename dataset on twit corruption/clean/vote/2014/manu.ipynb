```python
import pandas as pd
```

```python
df = pd.read_csv("vote_clean_2014.csv")
df
```

```python
#sort by polarity
df = df.sort_values(by=['polarity'], ascending=False)
df = df.reset_index(drop=True)
```

```python
df
```

```python
#remove polarity between 0.1 and 0.4
for i in range(0, 990):
    if (df.polarity.iloc[i] > 0.2) and (df.polarity.iloc[i] > 0.4):
        df = df.drop(df.index[i])

df = df.reset_index(drop=True)
df
```

```python
#percentage of positive tweets
ptweetsPercent = df[(df.polarity > 0)].shape[0]/df.shape[0]*100
ptweetsPercent

df.to_csv("vote_clean_2014m.csv", index=False)
```
