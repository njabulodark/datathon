```python
import pandas as pd
```

```python
df = pd.read_csv('2017data.csv')
df = df.dropna()
df = df[['Hashtags']]
df
```

```python
#create a list of hashtags
def make_list_of_hashtags(hash_list):
    hash_list = [x.split(',') for x in hash_list]
    return hash_list

df['Hashtags'] = make_list_of_hashtags(hash_list)
df
```

```python
df.Hashtags[17][0].replace('\'', '').replace('[', '').replace(']', '').replace(' ', '')

def remove_brackets(hash_list):
    hash_list = [x.replace('\'', '').replace('[', '').replace(']', '').replace(' ', '') for x in hash_list]
    return hash_list

df['Hashtags'] = df['Hashtags'].agg(remove_brackets)
df
```

```python
df.reset_index(drop=True, inplace=True)
#combine all the lists into one list
lis = []
for i in range(len(df["Hashtags"])):
    for j in range(len(df["Hashtags"][i])):
        lis.append(df["Hashtags"][i][j])
lis
```

```python
#make one list of hashtags
has_data = pd.DataFrame(lis, columns=['Hashtags'])
has_data
```
